```python
#Import Necessary Libraries:
#Import the required Python libraries for data analysis, such as Pandas, NumPy, Matplotlib, and Seaborn.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset:
#Load the Titanic dataset into a Pandas DataFrame.

df = pd.read_csv("C:/Users/Shajeer Waqar/Documents/CodSoft/archive/Titanic.csv")
```

```python
#Data Visualization:
#Create visualizations to gain insights into the data.

#Survival Rate:
sns.countplot(data=df, x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
#Passenger Class (Pclass) Distribution:
sns.countplot(data=df, x='Pclass')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
#Age Distribution:
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
#Gender Distribution:
sns.countplot(data=df, x='Sex')
plt.title('Gender Distribution')
plt.show()
```

```python
#Siblings/Spouses Aboard (SibSP) Distribution:
sns.countplot(data=df, x='SibSp')
plt.title('Siblings/Spouses Aboard Distribution')
plt.show()
```

```python
#Parents/Children Aboard (Parch) Distribution:
sns.countplot(data=df, x='Parch')
plt.title('Parents/Children Aboard Distribution')
plt.show()
```

```python
#Fare Distribution:
sns.histplot(data=df, x='Fare', kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
#Pclass vs. Survival:
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#Sex vs. Survival:
sns.barplot(data=df, x='Sex', y='Survived')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
#Age vs. Survival:
sns.histplot(data=df, x='Age', hue='Survived', bins=20, kde=True)
plt.title('Age vs. Survival')
plt.show()
```
